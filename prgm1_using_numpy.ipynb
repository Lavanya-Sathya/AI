```python
import numpy as np

x=np.array([0,0,1,1])
y=np.array([0,1,0,1])

a=np.logical_not(x)
b=np.logical_not(y)

print("Negation: ")
print(np.logical_not(a))
print(np.logical_not(b))
print("\n")

print("conjuction: ")
print(np.logical_and(x,y))
print("\n")

print("disjuction")
print(np.logical_or(x,y))
print("\n")

print("implication:")
print(np.logical_or(a,y))
print("\n")

print("Biconditional")
w=np.logical_or(a,y)
z=np.logical_or(b,x)
print(np.logical_and(w,z))
```
