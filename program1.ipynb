```python
def disjuction(a,b):
    if(a==1 & b==1):
        return 1
    else:
        return a+b
x=int(input('Enter value of x: '))
y=int(input('Enter value of y: '))
print("disjunction of ",x,"v",y," is ",x or y)
print("disjunction of ",x,"v",y," is ",disjuction(x,y))
```

```python
def implication(a,b):
    if(a==1 & b==0):
        return 0
    else:
        return 1
x=int(input('Enter value of x: '))
y=int(input('Enter value of y: '))
print("implication of ",x,"->",y," is ",implication(x,y))
```

```python
def res(a,b):
    if(a==1 & b==1):
        return 1
    elif(a==0 & b==0):
        return 1
    else:
        return 0
x=int(input('Enter value of x: '))
y=int(input('Enter value of y: '))
print("Biconditional of ",x,"<-->",y," is ",res(x,y))
```
