```python
import numpy as np

    
S=np.array([0,0,1,1])
T=np.array([0,1,0,1])

# not(s) , not(t)
X=np.logical_not(S)
Y=np.logical_not(T)

#not(s or t)
OR=np.logical_or(S,T)
a=np.logical_not(OR)

#(s and t)
b=np.logical_and(S,T)

#(t or not(t))
c=np.logical_or(T,Y)

#not(s<->s)
#(not(s) or s) and (s or not(s))

or1=np.logical_or(X,S)
or2=np.logical_or(S,X)
and1=np.logical_and(or1,or2)
d=np.logical_not(and1)

#not(s)->not(t)  here not(s) is x and not(t) is y
# so x->y can be reduced to not(x) or y
# not(x) is same as s
e=np.logical_or(S,Y)

    
print("not(S): ")
print(X)
print("not(T):")
print(Y)
print("a : ")
print(a)
print("b : ")
print(b)
print("c : ")
print(c)
print("d : ")
print(d)
print("e : ")
print(e)
print("\n")

if any(np.logical_and(a,b)):
    print("a entails b")
else:
    print("a does not entails b")
    

if any(np.logical_and(a,c)):
    print("a entails c")
else:
    print("a does not entails c")


if any(np.logical_and(a,d)):
    print("a entails d")
else:
    print("a does not entails d")

    
if any(np.logical_and(a,e)):
    print("a entails e")
else:
    print("a does not entails e")
```
